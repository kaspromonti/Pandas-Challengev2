### Heroes Of Pymoli Data Analysis

780 purchases were made by 576 unique buyers. 26.2% of our customers are repeat buyers. Males make up an overwhelming percentage of total sales with 84%, while Females and Others/Non-Disclosed make up the remaining 16%. However, Females and Others/Non-Disclosed outspend Males by a range of 46 - 56 cents per sale. Finally, 63% of all purchases are made by people between the ages of 15-24, with 20-24 year as our the greatest spenders making up 44.7% of that total. People over the age of 40 spend the least amount of money making up only 2% of total sales. The most popular and profitable item sold was Final Critc.

```python
# Import Dependencies and lod in csv file
import pandas as pd

# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
sales_df = pd.read_csv(file_to_load)
sales_df.head()
```

## Player Count

```python
# Calculating total number of "players" that make up sales data
player_count = len(sales_df["SN"].value_counts())

print(f'The total number of players is {player_count}.')
```

### Purchasing Analysis (Total)
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame


```python
# purchase_analysis_df
# Calculating Purchasing Analysis stats

unique_items = len(sales_df["Item ID"].unique())
avg_purchase = sales_df["Price"].mean()
total_purchases = sales_df["SN"].count()
total_rev = sales_df["Price"].sum()
                      
# Creating df to hold summary values
purchase_analysis_df = pd.DataFrame({"Total Player Count": [player_count],
                                  "Number of Unique Items": unique_items,
                                  "Average Purchase Price": avg_purchase,
                                  "Total Number of Purchases": total_purchases,
                                  "Total Revenue": total_rev})
purchase_analysis_df = purchase_analysis_df.set_index("Total Player Count")


purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)

# DISPLAYING PURCHASE ANALYSIS SUMMARY DATA
purchase_analysis_df
```

## Gender Demographics
* Percentage and Count of Male Players

* Percentage and Count of Female Players

* Percentage and Count of Other / Non-Disclosed

```python
# Calculating Gender Demographics 

# counting the unique values within the SN Column
gender_count = sales_df.groupby("Gender")["SN"].unique()

# counting unique values in each type of gender to return totals
female_count = len(gender_count["Female"])
male_count = len(gender_count["Male"])
other_count = len(gender_count["Other / Non-Disclosed"])

# Calculating Percentages

female_percentage = (female_count / player_count) * 100
male_percentage = (male_count / player_count) * 100
other_percentage = (other_count / player_count) * 100

gender_counts_df = pd.DataFrame({"Gender" : ["Female", "Male", "Other / NonDisclosed"],
                    "Count" : [female_count, male_count, other_count],
                    "% of Count" : [female_percentage, male_percentage, other_percentage]})

# formatting Percentages
gender_counts_df["% of Count"] = gender_counts_df["% of Count"].map("{:.2f}%".format)

# DISPLAYING GENDER DEMOGRAPHICS SUMMARY DATA
gender_counts_df
```


## Purchasing Analysis (Gender)
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

```python
# Calculating Purchasing Analysis based on total sales by gender
gender_demo_count = sales_df.groupby("Gender")["Gender"].count()
gender_demo_avg_price = sales_df.groupby("Gender")["Price"].mean()
gender_demo_total_rev = sales_df.groupby("Gender")["Price"].sum()
avg_spend_per_person = gender_demo_total_rev / (female_count, male_count, other_count)

gender_demo_df = pd.DataFrame({"Purchase Count" : gender_demo_count,
                             "Avg Purchase Price" : gender_demo_avg_price,
                              "Total Purchase Value" : gender_demo_total_rev,
                              "Avg Total by Person": avg_spend_per_person })

#Formatting Currency
gender_demo_df["Avg Purchase Price"] = gender_demo_df["Avg Purchase Price"].map("${:,.2f}".format)
gender_demo_df["Total Purchase Value"] = gender_demo_df["Total Purchase Value"].map("${:,.2f}".format)
gender_demo_df["Avg Total by Person"] = gender_demo_df["Avg Total by Person"].map("${:,.2f}".format)


#DISPLAYING PURCHASING ANALYSIS BY GENDERY TYPE SUMMARY
gender_demo_df
```

## Age Demographics

* Establish bins for age

* Categorize the existing players using the age bins. Hint: use pd.cut()

* Calculate the numbers and percentages by age group

* Create a summary data frame to hold the results

* Optional: round the percentage column to two decimal points

* Display Age Demographics Table

```python
# Creating new df that is specific to Age Demographics
# Pulling in only 3 columns from sales_df
age_df = sales_df.loc[:, ["Gender", "Age", "SN"]]
 
# Removing Duplicate values so that only the unique SNs are displayed
age_df = age_df.drop_duplicates()
```

```python
# Purchase Analysis by Age Demographic
# creating bins for age_df
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_df["Age Range"] = pd.cut(age_df["Age"], bins, labels=group_names, include_lowest=True)
age_df

count_by_age_range = age_df.groupby("Age Range")["Gender"].count()
percent_by_age_range = (count_by_age_range / player_count) * 100

age_summary_df = pd.DataFrame({"Total Count" : count_by_age_range,
                            "% By Age Range" : percent_by_age_range,
                            })
# formatting Percentages
age_summary_df["% By Age Range"] = age_summary_df["% By Age Range"].map("{:.2f}%".format)

#  DISPLAYING AGE DEMOGRAPHICS SUMMARY DATA
age_summary_df
```

# Purchasing Analysis (Age)

* Bin the purchase_data data frame by age

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

```python
# Purchase Analysis statistic calculations

# creating bins for age ranges
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Creating new column in sales_df by cutting "Age" 
sales_df["Age Range"] = pd.cut(sales_df["Age"], bins, labels=group_names, include_lowest=True)


# Calculating Age Demographics using grouping off sale_df
count_by_age = sales_df.groupby("Age Range")["Gender"].count()
avg_price_by_age = sales_df.groupby("Age Range")["Price"].mean()
percent_by_age = (sales_df.groupby("Age Range")["Price"].sum() / total_rev) * 100

total_rev_by_age = sales_df.groupby("Age Range")["Price"].sum()
avg_rev_by_SN = total_rev_by_age / count_by_age_range
```

```python
# PURCHASE ANALYSIS (AGE)
# Creating new df for Age Demographic Summary
purch_demobyage_df = pd.DataFrame({"Purchase Count" : count_by_age,
                             "% of Sales" : percent_by_age,
                            "Average Purchase Price" : avg_price_by_age,
                            "Total Purchase Value" : total_rev_by_age,
                            "Avg Total Purchase per Person": avg_rev_by_SN})

#Formatting Currency
purch_demobyage_df["Average Purchase Price"] = purch_demobyage_df["Average Purchase Price"].map("${:,.2f}".format)
purch_demobyage_df["Total Purchase Value"] = purch_demobyage_df["Total Purchase Value"].map("${:,.2f}".format)
purch_demobyage_df["Avg Total Purchase per Person"] = purch_demobyage_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

# formatting Percentages
purch_demobyage_df["% of Sales"] = purch_demobyage_df["% of Sales"].map("{:.2f}%".format)

# DISPLAYING PURCHASE ANALYSIS (AGE)
purch_demobyage_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below

* Create a summary data frame to hold the results

* Sort the total purchase value column in descending order

* Optional: give the displayed data cleaner formatting

* Display a preview of the summary data frame

```python
# TOPS SPENDERS ANALYSIS
# using Groupby and function to obtain Top Spender statistics
top_spender_count = sales_df.groupby("SN")["Price"].count()
top_spender_avg_price = sales_df.groupby("SN")["Price"].mean()
top_spender_total_rev = sales_df.groupby("SN")["Price"].sum()

top_spender_df = pd.DataFrame({
     "Purchase Count": top_spender_count,
    "Average Purchase Price":top_spender_avg_price,
    "Total Purchase Value":top_spender_total_rev
    })
#Sorting the df to show the top spenders 
top_spender_df=top_spender_df.sort_values(by="Total Purchase Value", ascending=False)

#Formatting Currency
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:,.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:,.2f}".format)

# DISPLAYING TOP 10 SPENDERS 
top_spender_df.head(10)
```

## Most Popular Items
* Retrieve the Item ID, Item Name, and Item Price columns

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

* Create a summary data frame to hold the results

* Sort the purchase count column in descending order

* Optional: give the displayed data cleaner formatting

* Display a preview of the summary data frame

```python
# Creating new df from sales_df for Popular Items Analysis
popular_df = sales_df.loc[:, ["Item Name","Item ID", "Price"]]
```

```python
# using Groupby and function to obtain Top Spender statistics
groupedby_popular = popular_df.groupby(["Item ID", "Item Name"])

# performing calculations base on the popular item groups
popular_count = groupedby_popular.count()["Price"]
popular_avg = groupedby_popular.mean()["Price"]
popular_sum = groupedby_popular.sum()["Price"]

popular_item_df = pd.DataFrame({
     "Total Item Count": popular_count,
    "Item Price":popular_avg ,
    "Total Purchase Value":popular_sum
    })
#Sorting the df to show the top selling intems 
popular_item_df = popular_item_df.sort_values(by="Total Item Count", ascending=False)

#Formatting Currency
popular_item_df["Item Price"] = popular_item_df["Item Price"].map("${:,.2f}".format)


# DISPLAYING TOP 10 MOST POPULAR ITEMS 
popular_item_df.head(20)
```
